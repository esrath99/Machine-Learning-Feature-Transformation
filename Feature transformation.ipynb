# Normal Techniques

```python
df2=pd.read_csv('supershops.csv')
df2.head()
```

```python
maximum = df2.Administration.max()
maximum
```

```python
minimum=df2.Administration.min()
minimum
```

```python
df2.Administration =  df2.Administration / maximum
```

```python
df2.head()
```

```python
df2.Administration.std()
```

```python
df2.Administration.var()
```

```python
df2.Administration.mean()
```
